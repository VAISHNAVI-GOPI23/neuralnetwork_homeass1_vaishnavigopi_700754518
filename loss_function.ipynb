```python
# üìå Loss Functions & Hyperparameter Tuning
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ Define true values (y_true) and model predictions (y_pred)
y_true = tf.constant([[0, 0, 1]], dtype=tf.float32)  # one-hot encoded
y_pred_1 = tf.constant([[0.1, 0.2, 0.7]], dtype=tf.float32)  # good prediction
y_pred_2 = tf.constant([[0.3, 0.3, 0.4]], dtype=tf.float32)  # worse prediction

# 2Ô∏è‚É£ Compute MSE and Categorical Cross-Entropy (CCE) Losses
mse_1 = tf.keras.losses.MeanSquaredError()(y_true, y_pred_1).numpy()
cce_1 = tf.keras.losses.CategoricalCrossentropy()(y_true, y_pred_1).numpy()

mse_2 = tf.keras.losses.MeanSquaredError()(y_true, y_pred_2).numpy()
cce_2 = tf.keras.losses.CategoricalCrossentropy()(y_true, y_pred_2).numpy()

# 3Ô∏è‚É£ Print loss values
print("=== Loss for Prediction 1 (closer to true) ===")
print("MSE:", mse_1)
print("Categorical Cross-Entropy:", cce_1)

print("\n=== Loss for Prediction 2 (worse) ===")
print("MSE:", mse_2)
print("Categorical Cross-Entropy:", cce_2)

# 4Ô∏è‚É£ Plot loss function values using Matplotlib
labels = ['Prediction 1', 'Prediction 2']
mse_values = [mse_1, mse_2]
cce_values = [cce_1, cce_2]

x = np.arange(len(labels))  # label locations
width = 0.35  # width of the bars

fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, mse_values, width, label='MSE')
bars2 = ax.bar(x + width/2, cce_values, width, label='Cross-Entropy')

# Labels and titles
ax.set_ylabel('Loss Value')
ax.set_title('Loss Function Comparison (Lower is Better)')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Display loss values on bars
for bar in bars1 + bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2.0, yval + 0.01, f'{yval:.4f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
